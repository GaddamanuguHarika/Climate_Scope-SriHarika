```python
import pandas as pd

filepath = 'GlobalWeatherRepository.csv'
df = pd.read_csv(filepath)
```

```python
print('Shape before cleaning:')
df.shape
print(df.head())
```

```python
str_cols = df.select_dtypes(include='object').columns
df[str_cols] = df[str_cols].apply(lambda col: col.str.strip())
```

```python
df['last_updated_datetime'] = pd.to_datetime(df['last_updated'], errors='coerce')

df['date'] = df['last_updated_datetime'].dt.date
df['time'] = df['last_updated_datetime'].dt.time

df['year'] = df['last_updated_datetime'].dt.year
df['month'] = df['last_updated_datetime'].dt.month
df['day'] = df['last_updated_datetime'].dt.day
df['hour'] = df['last_updated_datetime'].dt.hour
```

```python
df = df.drop_duplicates()

subset_cols = ['country', 'locationname', 'lastupdated']
if all(col in df.columns for col in subset_cols):
    df = df.drop_duplicates(subset=subset_cols, keep='last')
```

```python
num_cols = [
    'latitude', 'longitude', 'lastupdatedepoch',
    'temperaturecelsius', 'temperaturefahrenheit',
    'windmph', 'windkph', 'winddegree',
    'pressuremb', 'pressurein',
    'precipmm', 'precipin',
    'humidity', 'cloud',
    'feelslikecelsius', 'feelslikefahrenheit',
    'visibilitykm', 'visibilitymiles',
    'uvindex', 'gustmph', 'gustkph',
    'airqualityCarbonMonoxide', 'airqualityOzone',
    'airqualityNitrogendioxide', 'airqualitySulphurdioxide',
    'airqualityPM2.5', 'airqualityPM10',
    'airqualityus-epa-index', 'airqualitygb-defra-index',
    'moonillumination',
]
```

```python
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print('Shape after cleaning:')
df.shape
print(df.dtypes)
```

```python
clean_path = r'C:\springboard project.csv'
df.to_csv(clean_path, index=False)
print(f'Cleaned dataset saved as {clean_path}')
```
